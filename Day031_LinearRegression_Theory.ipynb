 Step 1 — What is Linear Regression?
- Definition (simple words): It’s a way to draw the “best straight line” through data points so we can predict outcomes.
- Goal: Find the relationship between input (X) and output (Y).

 Step 2 — Equation of Linear Regression
Y=\beta _0+\beta _1X+\epsilon 
- Y: Target (what we want to predict, e.g., salary)
- X: Feature (input, e.g., years of experience)
- \beta _0: Intercept (starting point when X=0)
- \beta _1: Slope (how much Y changes when X increases by 1)
- \epsilon : Error (difference between prediction and reality)
 Layman analogy: Imagine predicting house price from size. Every extra 100 sq ft adds ₹50,000 — that’s the slope. The base price when size = 0 is the intercept.

 Step 3 — Assumptions of Linear Regression
- Linearity: Relationship between X and Y is straight‑line.
- Independence: Each data point is separate, not linked.
- Equal variance (Homoscedasticity): Errors spread evenly.
- Normality: Errors follow a bell curve.

 Step 4 — Advantages
- Easy to understand and explain.
- Works well when data has a clear linear trend.

 Step 5 — Limitations
- Doesn’t work if data is curved/non‑linear.
- Sensitive to outliers (extreme values can distort the line).
- Assumes strict conditions.
