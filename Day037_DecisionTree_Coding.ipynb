```python
# pandas → handle tables
# numpy → handle numbers
# matplotlib/seaborn → visualize data
# sklearn → has decision tree models

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Example: Predict if a student passes (1) or fails (0)
# based on study hours and attendance.

data = {
    'StudyHours': [2,4,6,8,10,12,14,16],
    'Attendance': [60,65,70,75,80,85,90,95],
    'Pass': [0,0,0,1,1,1,1,1]  # 0 = Fail, 1 = Pass
}
df = pd.DataFrame(data)

df.head()
```

```python
# X = inputs (study hours + attendance)
# y = output (pass/fail)

X = df[['StudyHours','Attendance']]
y = df['Pass']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create a decision tree classifier
# max_depth = limit tree size (avoid overfitting)

model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X_train, y_train)
```

```python
# Plot the tree structure
plt.figure(figsize=(10,6))
plot_tree(model, feature_names=['StudyHours','Attendance'], class_names=['Fail','Pass'], filled=True)
plt.show()
```

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Predict for a student with 9 study hours and 80% attendance
sample = pd.DataFrame([[9,80]], columns=['StudyHours','Attendance'])
prediction = model.predict(sample)
print("Predicted outcome (1=Pass, 0=Fail):", prediction[0])
```
