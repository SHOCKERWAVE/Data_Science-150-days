```python
# Day 1: Variables & Data Types

# Storing a number in a variable
age = 22   # age is a box, 22 is inside it
print("Age:", age)

# Storing text in a variable
name = "Shubham"   # name is a box, "Shubham" is inside it
print("Name:", name)

# Storing decimal numbers
height = 5.9   # height is a float (decimal number)
print("Height:", height)

# Storing True/False values
is_student = True   # Boolean data type
print("Is Student:", is_student)

# Checking data types
print(type(age))        # int
print(type(height))     # float
print(type(name))       # str
print(type(is_student)) # bool
```
