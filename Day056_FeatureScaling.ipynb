 Step 1 — What is Feature Scaling?
- Definition: Adjusting the range of features so they are on a similar scale.
- Why: Some models (like Logistic Regression, SVM, KNN) are sensitive to feature magnitude.
- Layman analogy: Imagine a race where one runner measures distance in meters and another in kilometers. Without scaling, the kilometer runner looks much faster. Scaling ensures everyone runs on the same track.

 Step 2 — Methods of Feature Scaling
- Normalization (Min‑Max Scaling):
- Scales values between 0 and 1.
- Formula:
X'=\frac{X-X_{min}}{X_{max}-X_{min}}- Standardization (Z‑Score Scaling):
- Centers data around mean = 0, standard deviation = 1.
- Formula:
X'=\frac{X-\mu }{\sigma }

```python
# Step 3
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load dataset
iris = load_iris()
X = iris.data   # features (petal length, width, etc.)

# Min-Max Scaling (Normalization)
minmax_scaler = MinMaxScaler()
X_minmax = minmax_scaler.fit_transform(X)
print("First 5 rows after Min-Max Scaling:\n", X_minmax[:5])

# Standardization (Z-Score Scaling)
std_scaler = StandardScaler()
X_std = std_scaler.fit_transform(X)
print("\nFirst 5 rows after Standardization:\n", X_std[:5])
```
