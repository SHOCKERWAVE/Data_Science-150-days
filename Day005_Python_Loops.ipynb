```python
# Day 5: Loops

# Example 1: Print numbers from 1 to 5
for i in range(1, 6):   # range(1,6) means 1 to 5
    print(i)

# Example 2: Loop through a list
fruits = ["apple", "banana", "mango"]
for fruit in fruits:
    print("Fruit:", fruit)
```

```python
# Example 1: Print numbers until 5
count = 1
while count <= 5:   # keep running until count is 5
    print("Count:", count)
    count += 1   # increase count by 1 each time

# Example 2: Stop when condition fails
num = 10
while num > 0:
    print("Number:", num)
    num -= 2   # decrease by 2 each time
```
