```python
# for = Repeat a fixed number of times

#Print number 1 to 5
for i in range(1, 6):
    print(i)
```

```python
# Loop through a list
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

```python
# while = repeat until condition becomes false
count = 1

while count <=5:
    print("Count is:", count)
    count += 1       # increase count by 1
```

```python
# break = stop the loop early
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

```python
# continue = skip one step
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

    # Day 5 - Python Loops (For & While)

    Learned:
    -For loop = repeat a fixed number of times
    -While loop = repeat until condition becomes false
    -break = stop the loop early
    -continue = skip one step
