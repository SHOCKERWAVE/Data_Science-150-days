```python
# Creating a class called Person
#Think of this as a blueprint for making 'person' objects
class Person:
    
    # __init__ is a special method called that runs automatically
    # It sets up the data (attributes) for each object
    def __init__(self, name, age):
        self.name = name    # Storing the name inside the object
        self.age = age      # Storing the age inside the object

    # A method = an action the object can perform
    def introduce(self):
        print(f"My name is {self.name} and I am {self.age} years old.")

# Creating an object from the Person class
p1 = Person("Shubham", 22)  # p1 is now a real 'person' object
p1.introduce()              # Calling the introduce method
```

```python
# A class to represent a car
class Car:

    # Setting up the car's basic information
    def __init__(self, brand, model):
        self.brand = brand    # storing brand
        self.model = model    # storing model

    # A method to show car details
    def details(self):
        print(f"Car: {self.brand} {self.model}")

# Creating an object of Car
car1 = Car("Honda", "City")
car1.details()
```

```python
# A simple calculator class
class Calculator:

    # Method to add two numbers
    def add(self, a, b):
        return a + b
    
# Creating an object of Calculator
calc = Calculator()
print(calc.add(10, 5))  #calling the add method
```
