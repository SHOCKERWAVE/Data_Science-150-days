# TensorFlow Basics

**Goal:** Learn the basics of TensorFlow and build a simple model.


**TensorFlow** is a popular library for deep learning.  
It helps us build, train, and evaluate neural networks easily.

---

## Why TensorFlow?
- Handles complex math automatically.
- Works with both CPUs and GPUs.
- Provides ready-made tools for building models.

---

## Layman Explanation
Think of TensorFlow as a **toolbox for building robots**:
- You don’t need to design every screw or wire.
- The toolbox already has parts (layers, optimizers, loss functions).
- You just assemble them to make your robot (neural network).


```python
import tensorflow as tf

# Check TensorFlow version
print("TensorFlow version:", tf.__version__)

# Simple constant
hello = tf.constant("Hello, TensorFlow!")
print(hello.numpy())
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# Build a simple model
model = Sequential([
    Input(shape=(10,)),   # input layer with 10 features
    Dense(8, activation='relu'),  # hidden layer
    Dense(1, activation='sigmoid') # output layer
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

print(model.summary())
```

- We created a simple neural network with TensorFlow.
- Input layer: accepts 10 features.
- Hidden layer: processes data with ReLU activation.
- Output layer: gives a probability (sigmoid).

Layman:
It’s like building a small robot:
- Input = sensors (data).
- Hidden = brain (processing).
- Output = decision (yes/no).
