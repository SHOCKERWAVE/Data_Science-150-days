```python
import matplotlib.pyplot as plt

# Example data
days = [1, 2, 3, 4, 5]
sales = [10, 20, 15, 25, 30]

# Line plot
plt.plot(days, sales)   # x = days, y = sales
plt.title("Sales Over Days")   # chart title
plt.xlabel("Days")             # x-axis label
plt.ylabel("Sales")            # y-axis label
plt.show()
```

```python
# Example data
students = ["Shubham", "Rahul", "Priya"]
marks = [85, 90, 78]

# Bar plot
plt.bar(students, marks)
plt.title("Student Marks")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
# Line plot with style
plt.plot(days, sales, color="red", marker="o", linestyle="--")
plt.title("Styled Sales Plot")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.show()
```
