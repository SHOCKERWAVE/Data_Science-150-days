# Forward Propagation

**Goal:** Understand how data flows through a neural network step by step.


Forward propagation is the process of passing input data through the network, layer by layer, until we get an output.

---

## Steps
1. Input data enters the first layer.
2. Each neuron multiplies inputs by weights and adds bias.
3. Activation functions decide whether to pass the signal forward.
4. The output layer produces the final prediction.

---

## Layman Explanation
Think of a relay race:
- The first runner (input layer) starts with the baton (data).
- Each runner (hidden layer) adds their own effort (weights + bias).
- The coach (activation function) decides if the runner should continue.
- The last runner (output layer) finishes the race and gives the result (prediction).


```python
import numpy as np

# Input data
X = np.array([0.5, 0.8])

# Weights and bias
W1 = np.array([[0.2, 0.4], [0.6, 0.9]])
b1 = np.array([0.1, 0.2])

W2 = np.array([[0.3], [0.5]])
b2 = np.array([0.05])

# Activation function (ReLU)
def relu(x):
    return np.maximum(0, x)

# Forward propagation
hidden_input = np.dot(X, W1) + b1
hidden_output = relu(hidden_input)

final_input = np.dot(hidden_output, W2) + b2
final_output = relu(final_input)

print("Final Output:", final_output)
```

- Input data goes into the first layer.
- Each neuron multiplies inputs by weights and adds bias.
- ReLU activation decides if the signal should pass forward.
- The final layer produces the output.

Layman:
It’s like solving a math problem step by step:
- First, you calculate with weights.
- Then, you add a small hint (bias).
- Finally, you decide if the answer is useful (activation).
- The last step gives the final result.
