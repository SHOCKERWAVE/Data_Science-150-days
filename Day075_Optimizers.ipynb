# Optimizers (SGD, Adam)

**Goal:** Understand how optimizers adjust weights to make neural networks learn efficiently.


An **optimizer** is a method used to update the weights of a neural network during training.  
It decides how much to change the weights after each mistake.

---

## Why Optimizers?
- Reduce errors faster.
- Prevent the network from getting stuck in bad solutions.
- Make training more efficient.

---

## Layman Explanation
Imagine a student practicing math:
- After each mistake, they adjust their method.
- Optimizers are like different **study strategies**:
  - Some students change answers slowly and carefully (SGD).
  - Others adjust quickly and smartly using past experience (Adam).


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# Simple model with proper Input layer
model = Sequential([
    Input(shape=(10,)),   # Correct way to define input
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile with SGD
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Compile with Adam
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

### 1. SGD (Stochastic Gradient Descent)
- Updates weights step by step.
- Simple and reliable.
- Layman: Like a student improving slowly, one small correction at a time.

### 2. Adam (Adaptive Moment Estimation)
- Uses past knowledge to adjust faster.
- More efficient for complex problems.
- Layman: Like a smart student who remembers past mistakes and learns quicker.
