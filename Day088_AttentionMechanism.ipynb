# Attention Mechanism

**Goal:** Learn how attention helps models focus on important parts of input sequences.


The **attention mechanism** allows a model to assign different importance (weights) to different parts of the input.  
Instead of treating all words equally, the model focuses more on the relevant ones.

---

## Why Attention?
- Improves context understanding.
- Helps with long sentences or sequences.
- Forms the foundation of transformer models.

---

## Layman Explanation
Imagine a student listening to a lecture:
- They don’t remember every single word equally.
- They highlight key points in their notebook.
- Attention = the student’s highlighter, marking important words or ideas.


```python
import tensorflow as tf
from tensorflow.keras.layers import Attention, Input
from tensorflow.keras.models import Model
import numpy as np

# Dummy input: 2 sequences of length 5 with 4 features each
query = np.random.random((1, 5, 4)).astype(np.float32)
value = np.random.random((1, 5, 4)).astype(np.float32)

# Attention layer
inputs_q = Input(shape=(5, 4))
inputs_v = Input(shape=(5, 4))
attn_out = Attention()([inputs_q, inputs_v])

model = Model(inputs=[inputs_q, inputs_v], outputs=attn_out)

# Run attention
output = model.predict([query, value])
print("Attention output shape:", output.shape)
```

- Query = what we are focusing on.
- Value = the information available.
- Attention layer = decides which parts of the value are most important for the query.

Layman:
It’s like asking a question in class:
- Query = the question.
- Value = the lecture notes.
- Attention = the student highlights the parts of notes that answer the question.
