```python
# sklearn → has machine learning models and evaluation tools
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# Load the famous Iris dataset (flowers classification)
iris = load_iris()
X = iris.data   # features (petal length, width, etc.)
y = iris.target # labels (flower species)
```

```python
# Create a Decision Tree model
dt_model = DecisionTreeClassifier(random_state=42)

# Perform 5-fold cross validation
# Meaning: split data into 5 parts → train on 4, test on 1 → repeat 5 times
scores_dt = cross_val_score(dt_model, X, y, cv=5)

# Print average accuracy across folds
print("Decision Tree CV Accuracy:", scores_dt.mean())
```

```python
# Create a Logistic Regression model
lr_model = LogisticRegression(max_iter=200)

# Perform 5-fold cross validation
scores_lr = cross_val_score(lr_model, X, y, cv=5)

print("Logistic Regression CV Accuracy:", scores_lr.mean())
```
