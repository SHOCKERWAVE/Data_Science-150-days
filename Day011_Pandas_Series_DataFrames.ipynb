```python
import pandas as pd   # pd is the short name for Pandas
```

```python
# Create a Series (like one column)
numbers = pd.Series([10, 20, 30, 40])
print("Series:\n", numbers)

# Access by index
print("First element:", numbers[0])
```

```python
# Create a DataFrame (like a table)
data = {
    "Name": ["Shubham", "Rahul", "Priya"],
    "Age": [22, 25, 23],
    "City": ["Prayagraj", "Delhi", "Mumbai"]
}

df = pd.DataFrame(data)
print("DataFrame:\n", df)

# Access a column
print("Names:\n", df["Name"])

# Access a row
print("First row:\n", df.iloc[0])
```

```python
# Get shape (rows, columns)
print("Shape:", df.shape)

# Get column names
print("Columns:", df.columns)

# Describe numerical data
print("Statistics:\n", df.describe())
```
