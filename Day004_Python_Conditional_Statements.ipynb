```python
# Day 4: Conditional Statements

# Example 1: Simple if
age = 18
if age >= 18:
    print("You are an adult.")   # runs only if condition is true

# Example 2: if-else
marks = 40
if marks >= 50:
    print("Pass")
else:
    print("Fail")   # runs if condition is false

# Example 3: if-elif-else
number = 0
if number > 0:
    print("Positive number")
elif number < 0:
    print("Negative number")
else:
    print("Zero")
```
