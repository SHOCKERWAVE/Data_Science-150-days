# MNIST Classifier

**Goal:** Build a neural network to classify handwritten digits using the MNIST dataset.


The MNIST dataset contains 70,000 images of handwritten digits (0–9).  
We will build a neural network that learns to recognize these digits.

---

## Steps
1. Load and preprocess the dataset.
2. Build a neural network model.
3. Compile the model with optimizer and loss.
4. Train the model on training data.
5. Evaluate the model on test data.

---

## Layman Explanation
Think of this as teaching a robot to **read human handwriting**:
- Step 1: Show the robot lots of handwritten numbers.
- Step 2: Build its brain (neural network).
- Step 3: Decide how it will learn (optimizer + loss).
- Step 4: Give it practice sessions (training).
- Step 5: Test it with new handwriting (evaluation).


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist

# Step 1: Load dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize data (scale pixel values between 0 and 1)
X_train = X_train / 255.0
X_test = X_test / 255.0

# Step 2: Build model
model = Sequential([
    Flatten(input_shape=(28, 28)),   # flatten 28x28 image into 784 features
    Dense(128, activation='relu'),   # hidden layer
    Dense(10, activation='softmax')  # output layer (10 classes)
])

# Step 3: Compile model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

print(model.summary())
```

- Flatten layer: turns each 28x28 image into a list of numbers.
- Hidden layer: processes the numbers with ReLU activation.
- Output layer: predicts one of 10 digits using softmax.

Layman:
It’s like teaching a robot:
- Flatten = robot scans the whole page into a list.
- Hidden = robot thinks about the features.
- Output = robot decides which digit it is.
