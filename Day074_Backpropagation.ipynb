# Backpropagation

**Goal:** Learn how a neural network adjusts its weights by correcting mistakes.


Backpropagation is the process of sending errors backward through the network to update weights.

---

## Steps
1. Forward propagation → make a prediction.
2. Compare prediction with actual result → calculate error (loss).
3. Send error backward through the network.
4. Adjust weights and biases to reduce future errors.

---

## Layman Explanation
Imagine a student solving math problems:
- First, they try an answer (forward propagation).
- Teacher checks and says "Wrong, you’re off by 2 marks" (error).
- The student reviews each step to see where they went wrong (backpropagation).
- Next time, they adjust their method to avoid the same mistake (weight update).

This cycle continues until the student gets better and better.


```python
import numpy as np

# Input and true output
X = np.array([[0.5, 0.8]])
y_true = np.array([[1]])

# Weights
W1 = np.random.rand(2, 2)
W2 = np.random.rand(2, 1)

# Activation function (Sigmoid)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Forward propagation
hidden_input = np.dot(X, W1)
hidden_output = sigmoid(hidden_input)

final_input = np.dot(hidden_output, W2)
final_output = sigmoid(final_input)

# Error
error = y_true - final_output

# Backpropagation
d_final = error * sigmoid_derivative(final_output)
d_hidden = d_final.dot(W2.T) * sigmoid_derivative(hidden_output)

# Update weights
W2 += hidden_output.T.dot(d_final) * 0.1
W1 += X.T.dot(d_hidden) * 0.1

print("Updated Weights W1:", W1)
print("Updated Weights W2:", W2)
```

- The network makes a guess (forward propagation).
- We calculate the error (difference between guess and truth).
- Error is sent backward (backpropagation).
- We adjust weights slightly to improve the next guess.

Layman:
It’s like practicing math problems:
- Try → Check → Find mistake → Correct → Try again.
