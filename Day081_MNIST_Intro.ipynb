# MNIST Dataset Introduction

**Goal:** Learn about the MNIST dataset and prepare it for training neural networks.


The **MNIST dataset** contains 70,000 images of handwritten digits (0–9).  
Each image is 28x28 pixels in grayscale.

---

## Why MNIST?
- Simple and widely used for beginners.
- Helps test neural networks on image recognition.
- Acts as a "Hello World" dataset for deep learning.

---

## Layman Explanation
Think of MNIST as a **school exam of handwriting digits**:
- Each image is like a student writing a number.
- The neural network’s job is to read the handwriting correctly.
- Training on MNIST is like teaching a robot to recognize numbers written by different people.


```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist

# Load dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print("Training data shape:", X_train.shape)
print("Testing data shape:", X_test.shape)
```

```python
import matplotlib.pyplot as plt

# Show first 5 images
for i in range(5):
    plt.imshow(X_train[i], cmap='gray')
    plt.title(f"Label: {y_train[i]}")
    plt.show()
```

- We loaded the MNIST dataset (handwritten digits).
- Training set = 60,000 images.
- Testing set = 10,000 images.
- Each image is 28x28 pixels.
- We displayed a few sample digits with their labels.

Layman:
It’s like opening a notebook full of handwritten numbers.
- Training pages = practice questions.
- Testing pages = exam questions.
- Each page has a small picture of a digit.
