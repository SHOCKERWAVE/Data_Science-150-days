# Train MNIST Classifier

**Goal:** Train the neural network on the MNIST dataset and observe learning progress.


Training a neural network means showing it examples (images + labels),  
letting it guess, checking mistakes, and correcting weights until it improves.

---

## Steps
1. Feed training images into the model.
2. Model predicts digits.
3. Compare predictions with actual labels.
4. Adjust weights using backpropagation + optimizer.
5. Repeat for multiple rounds (epochs).

---

## Layman Explanation
Think of this as a **student practicing handwriting recognition**:
- The student looks at thousands of digit images.
- They guess the number.
- Teacher corrects them when wrong.
- After many practice rounds, the student gets better at recognizing digits.


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist

# Load dataset
(X_train, y_train), (X_test, y_test) = mnist.load_data()

# Normalize data
X_train = X_train / 255.0
X_test = X_test / 255.0

# Build model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

# Compile model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train model
history = model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.1, verbose=1)
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title('Training Progress')
plt.show()
```

- We trained the model for 5 rounds (epochs).
- Each round, the model saw all training images and improved.
- We plotted accuracy to see progress.
- Validation accuracy shows how well the model performs on unseen data.

Layman:
It’s like giving a student 5 practice exams:
- Each exam helps them learn better.
- We track their scores to see improvement.
- Validation = checking with new questions to ensure they’re not just memorizing.
