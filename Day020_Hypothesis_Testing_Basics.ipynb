```python
import numpy as np
from scipy import stats

# Day 20: Hypothesis Testing Basics

# Example: Test if average height = 170 cm
data = [172, 168, 171, 169, 173, 170, 174]

# Null hypothesis H0: mean = 170
# Alternative hypothesis H1: mean ≠ 170

# Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(data, 170)

print("T-statistic:", t_stat)
print("P-value:", p_value)

# Decision
alpha = 0.05
if p_value < alpha:
    print("Reject H0 → Mean is significantly different from 170")
else:
    print("Fail to reject H0 → Mean is not significantly different from 170")
```
