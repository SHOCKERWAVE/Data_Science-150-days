```python
# Day 2: Lists & Tuples

# A list stores multiple values
fruits = ["apple", "banana", "mango"]
print("Fruits list:", fruits)

# Access items by index (starts at 0)
print("First fruit:", fruits[0])

# Change an item
fruits[1] = "orange"
print("After changing banana to orange:", fruits)

# Add a new item
fruits.append("grapes")
print("After adding grapes:", fruits)

# Remove an item
fruits.remove("apple")
print("After removing apple:", fruits)
```

```python
# A tuple stores multiple values but cannot be changed
colors = ("red", "green", "blue")
print("Colors tuple:", colors)

# Access items
print("First color:", colors[0])

# Tuples are immutable → cannot change values
colors[0] = "yellow"  # This will give an error
```
