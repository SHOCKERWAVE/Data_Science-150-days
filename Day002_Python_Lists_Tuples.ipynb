```python
# Day 2: Lists & Tupes

# A list = a flexible bag (you can add/remove/change items)
fruits = ["apple", "banana", "cherry"]
print("Fruits list:", fruits)

# Add a new fruit
fruits.append("orange")
print("After adding orange:", fruits)

# change an item
fruits[1] = "kiwi"
print("After changing banana to kiwi:", fruits)

# Remove an item
fruits.remove("cherry")
print("After removing cherry:", fruits)
```

```python
# A tuple = a sealed bag (you cannot change items)
coordinates = (25.3, 40.7)
print("Coordinates:", coordinates)

# Try changing a tuple item (this will give an error)
coordinates[0] = 30.0
```

    .Lists are mutable (you can modify them.)
    .Tuples are immutable (you cannot modify them.)
