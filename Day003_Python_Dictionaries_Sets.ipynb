```python
# Day 3: Dictionaries & Sets

# Creating a dictionary
student = {
    "name": "Shubham",
    "age": 22,
    "city": "Prayagraj"
}
print("Student dictionary:", student)

# Accessing values using keys
print("Name:", student["name"])
print("Age:", student["age"])

# Adding a new key-value pair
student["marks"] = 95
print("After adding marks:", student)

# Changing a value
student["city"] = "Bangalore"
print("After changing city:", student)

# Removing a key-value pair
del student["age"]
print("After removing age:", student)
```

```python
# Creating a set
numbers = {1, 2, 3, 4, 4, 5}   # duplicate 4 will be removed
print("Numbers set:", numbers)

# Adding an item
numbers.add(6)
print("After adding 6:", numbers)

# Removing an item
numbers.remove(2)
print("After removing 2:", numbers)

# Sets do not allow duplicates
numbers.add(3)   # 3 already exists, so no change
print("After trying to add duplicate 3:", numbers)
```
