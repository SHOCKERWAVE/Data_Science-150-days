```python
# Day 3: Dictionaries & Sets

# A dictionary = key -> value pairs
student = {
    "name": "Shubham",
    "age": 22,
    "city": "Prayagraj"
}

print("Name:", student)

# Access a value using its key
print("Name:", student["name"])

# Add a new key-value pair
student["course"] = "Data Science"
print("After adding course:", student)

# Change a value
student['age'] = 23
print("After updating age:", student)

# Remove a keuy-value pair
student.pop('city')
print("After removing city:", student)
```

```python
# A set = collection of unique items
numbers = {1, 2, 3, 4, 4, 5}
print("Set with duplicates removed:", numbers)

# Add an item
numbers.add(6)
print("After adding 6:", numbers)

# Remove an item
numbers.remove(2)
print("After removing 2:", numbers)
```
