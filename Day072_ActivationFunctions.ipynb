# Activation Functions

**Goal:** Learn what activation functions are and how they help neural networks make decisions.


An **activation function** decides whether a neuron should "fire" (pass information forward) or stay quiet.  
Without activation functions, a neural network would just be a big calculator doing linear math.

---

## Why Activation Functions?
- Add **non-linearity** → helps the network learn complex patterns.
- Decide if a signal is strong enough to pass forward.
- Allow networks to solve problems like image recognition, language understanding, etc.

---

## Layman Explanation
Think of neurons as students:
- Each student hears an idea (input).
- The activation function decides: "Is this idea worth sharing?"
- If yes, the student speaks up (output). If not, they stay quiet.

This way, only useful ideas move forward.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# ReLU
def relu(x):
    return np.maximum(0, x)

# Tanh
def tanh(x):
    return np.tanh(x)

# Plot them
x = np.linspace(-5, 5, 100)

plt.figure(figsize=(12,4))

plt.subplot(1,3,1)
plt.plot(x, sigmoid(x))
plt.title("Sigmoid")

plt.subplot(1,3,2)
plt.plot(x, relu(x))
plt.title("ReLU")

plt.subplot(1,3,3)
plt.plot(x, tanh(x))
plt.title("Tanh")

plt.show()
```

### 1. Sigmoid
- Output between 0 and 1.
- Good for probabilities.
- Layman: Like a student saying "I’m 70% sure this is correct."

### 2. ReLU (Rectified Linear Unit)
- Outputs 0 if input is negative, otherwise passes it forward.
- Fast and widely used.
- Layman: Student only speaks if the idea is positive; stays silent otherwise.

### 3. Tanh
- Output between -1 and 1.
- Stronger than sigmoid for centered data.
- Layman: Student can say "I strongly disagree (-1)" or "I strongly agree (+1)."
