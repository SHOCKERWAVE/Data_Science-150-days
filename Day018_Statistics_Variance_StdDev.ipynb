```python
import statistics as stats

# Day 18: Standard Deviation & Variance

numbers = [10, 20, 30, 40, 50]

# Variance
variance_value = stats.variance(numbers)
print("Variance:", variance_value)

# Standard Deviation
std_dev_value = stats.stdev(numbers)
print("Standard Deviation:", std_dev_value)
```

```python
# Mean
mean_value = sum(numbers) / len(numbers)

# Variance manually
variance_manual = sum((x - mean_value) ** 2 for x in numbers) / (len(numbers) - 1)
print("Manual Variance:", variance_manual)

# Standard Deviation manually
std_dev_manual = variance_manual ** 0.5
print("Manual Standard Deviation:", std_dev_manual)
```
