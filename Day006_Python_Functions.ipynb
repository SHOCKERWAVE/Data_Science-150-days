```python
# Basic Function
def greet():
    print("Hello Shubham")

greet()
```

```python
# Function with Input 
def greet(name):
    print("Hello", name)

greet("Shubham")
greet("Rahul")
```

```python
# Function with Output (Return Value)
def add(a, b):
    return a + b

result = add(5, 3)
print("Result:", result)
```

```python
# Function with Default value
def welcome(city="Prayagraj"):
    print("Welcome to", city)

welcome()
welcome("Delhi")
```
