```python
# Day 6: Functions

# Defining a function
def greet():
    print("Hello, welcome to Day 6!")

# Calling the function
greet()
```

```python
# Function that takes input
def greet_user(name):
    print("Hello,", name)

greet_user("Shubham")
greet_user("Rahul")
```

```python
# Function that returns output
def add_numbers(a, b):
    return a + b

result = add_numbers(10, 5)
print("Sum:", result)
```

```python
# Function with default value
def introduce(name, city="Prayagraj"):
    print(f"My name is {name} and I live in {city}")

introduce("Shubham")          # uses default city
introduce("Rahul", "Bangalore")  # overrides default
```
