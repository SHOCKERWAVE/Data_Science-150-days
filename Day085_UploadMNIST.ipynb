# Upload MNIST Project to GitHub

**Goal:** Save, document, and upload the MNIST classifier project to GitHub.


Uploading to GitHub makes your project visible to recruiters and shows your progress.  
A clean README helps others understand your work quickly.

---

## Steps
1. Save your notebook and outputs.
2. Write a README.md with project summary.
3. Organize files into folders.
4. Commit changes.
5. Push to GitHub.

---

## Layman Explanation
Think of GitHub as your **online showcase**:
- Your project = a school assignment.
- README = cover page explaining what you did.
- Uploading = submitting your assignment to the teacher (recruiter).


# Initialize Git (if not already done)
git init

# Add files
git add Day081_MNIST_Intro.ipynb Day082_MNIST_Classifier.ipynb Day083_TrainMNIST.ipynb Day084_EvaluateMNIST.ipynb README.md

# Commit changes
git commit -m "MNIST Project: Intro, Classifier, Training, Evaluation"

# Push to GitHub
git branch -M main
git remote add origin https://github.com/yourusername/MNIST-Classifier.git
git push -u origin main
