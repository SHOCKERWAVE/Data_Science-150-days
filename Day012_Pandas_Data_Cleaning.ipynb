```python
import pandas as pd

# Sample messy dataset
data = {
    "Name": ["Shubham", "Rahul", "Priya", "Rahul"],
    "Age": [22, 25, None, 25],
    "City": ["Prayagraj", "Delhi", "Mumbai", "Delhi"]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

```python
# Remove duplicate rows
df_no_duplicates = df.drop_duplicates()
print("\nAfter removing duplicates:\n", df_no_duplicates)
```

```python
# Fill missing values with default
df_fill = df.fillna(0)
print("\nAfter filling missing values:\n", df_fill)

# Drop rows with missing values
df_drop = df.dropna()
print("\nAfter dropping missing values:\n", df_drop)
```

```python
# Rename column names
df_renamed = df.rename(columns={"Name": "Full_Name", "Age": "Years"})
print("\nAfter renaming columns:\n", df_renamed)
```

```python
# Convert Age column to integer (fill missing first)
df["Age"] = df["Age"].fillna(0).astype(int)
print("\nAfter converting Age to int:\n", df)
```
