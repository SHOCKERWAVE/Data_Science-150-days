 Step 1 — What is a Decision Tree?
- Definition (simple words): A decision tree is like a flowchart that splits data into branches based on conditions.
- Each branch represents a decision rule, and each leaf represents an outcome.
- Layman analogy: Imagine deciding whether to buy a laptop.
- First question: Is budget > ₹50,000?
- Yes → Go to branch “High‑end laptops.”
- No → Go to branch “Budget laptops.”
- Next question: Do you need gaming?
- Yes → Gaming laptop.
- No → Office laptop.

 Step 2 — Structure of a Decision Tree
- Root Node: First question/condition.
- Branches: Possible answers (Yes/No, or ranges).
- Leaf Nodes: Final decision or prediction.

 Step 3 — Splitting Criteria
Decision trees use math to decide the best split:
- Gini Index → Measures impurity.
- Entropy / Information Gain → Measures how much uncertainty is reduced.
- Variance Reduction → Used for regression trees.

 Step 4 — Advantages
- Easy to understand and visualize.
- Works for both classification (Yes/No) and regression (numbers).
- Handles categorical and numerical data.

 Step 5 — Limitations
- Can overfit (too many branches).
- Sensitive to small changes in data.
- Not always the most accurate compared to ensemble methods.
