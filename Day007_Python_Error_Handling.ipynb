```python
# Day 7: Error Handling

# Example: Division by zero
try:
    num = 10 / 0   # This will cause an error
except ZeroDivisionError:
    print("You cannot divide by zero!")   # Friendly message instead of crash
```

```python
try:
    x = int("hello")   # Cannot convert text to number
except Exception as e:
    print("Error happened:", e)   # Prints the error message
```

```python
try:
    num = int("20")   # This works
except:
    print("Something went wrong!")
else:
    print("Conversion successful:", num)   # Runs if no error
```

```python
try:
    file = open("abc.txt", "r")   # File may not exist
except:
    print("File not found!")
finally:
    print("This will run no matter what.")   # Always runs
```
