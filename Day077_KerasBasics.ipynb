# Keras Basics

**Goal:** Learn how to use Keras to build and train simple neural networks.


**Keras** is a high-level API built into TensorFlow.  
It allows us to build neural networks with just a few lines of code.

---

## Why Keras?
- Simple and beginner-friendly.
- Provides ready-made layers (Dense, Conv2D, etc.).
- Easy to train and evaluate models.

---

## Layman Explanation
Think of Keras as a **Lego kit for neural networks**:
- Each block (layer) is already designed.
- You just stack blocks together to build your model.
- No need to carve wood or metal — just assemble and play.


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

# Build a simple model
model = Sequential([
    Input(shape=(10,)),   # input layer with 10 features
    Dense(8, activation='relu'),  # hidden layer
    Dense(1, activation='sigmoid') # output layer
])

# Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

print(model.summary())
```

```python
import numpy as np

# Dummy dataset
X = np.random.rand(100, 10)   # 100 samples, 10 features
y = np.random.randint(0, 2, (100, 1))  # binary labels

# Train the model
history = model.fit(X, y, epochs=5, batch_size=10, verbose=1)
```

- We built a simple model with Keras.
- Input layer: accepts 10 features.
- Hidden layer: processes data with ReLU activation.
- Output layer: predicts yes/no with sigmoid.
- We trained the model on dummy data for 5 rounds (epochs).

It’s like teaching a robot:
- Input = robot’s sensors.
