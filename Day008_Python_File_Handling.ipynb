```python
# Day 8: File Handling

# Open a file in write mode
file = open("example.txt", "w")   # "w" = write
file.write("Hello, this is Day 8!")   # write text into file
file.close()   # always close the file
```

```python
# Open the file in read mode
file = open("example.txt", "r")   # "r" = read
content = file.read()   # read all content
print("File content:", content)
file.close()
```

```python
# Open the file in append mode
file = open("example.txt", "a")   # "a" = append
file.write("\nAdding another line.")   # add new line
file.close()
```

```python
# 'with' automatically closes the file
with open("example.txt", "r") as f:
    print("File content using with:", f.read())
```
