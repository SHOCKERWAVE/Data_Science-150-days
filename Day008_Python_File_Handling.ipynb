```python
# Write text into a file
file = open("notes.txt", "w")      # 'w' = write mode
file.write("Hello , this is Shubham.\nWelcome to Day 8!")
file.close()
```

```python
# Reading from a File
file = open("notes.txt", "r")      # 'r' = read mode
content = file.read()
print(content)
file.close()
```

```python
file = open("notes.txt", "a")      # 'a' = append mode
file.write("\nThis is an extra line added later.")
file.close()
```

```python
# with automatically closes the file
with open("notes.txt", 'r') as file:
    print(file.read())
```
