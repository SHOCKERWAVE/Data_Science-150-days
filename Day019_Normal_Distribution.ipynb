```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Day 19: Normal Distribution

# Generate random data following normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)
# loc = mean, scale = standard deviation, size = number of samples

# Plot histogram
sns.histplot(data, bins=30, kde=True)
plt.title("Normal Distribution (mean=50, sd=10)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
# Mean = 100, Standard Deviation = 15
data2 = np.random.normal(loc=100, scale=15, size=1000)

sns.histplot(data2, bins=30, kde=True)
plt.title("Normal Distribution (mean=100, sd=15)")
plt.show()
```
