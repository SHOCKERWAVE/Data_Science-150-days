# Build a Simple Neural Network

**Goal:** Use Keras to build and train a simple neural network on dummy data.


A **neural network** is built by stacking layers of neurons.  
Each layer transforms the input data step by step until we get an output.

---

## Steps
1. Define the model structure (layers).
2. Compile the model (choose optimizer, loss, metrics).
3. Train the model on data.
4. Evaluate the modelвЂ™s performance.

---

## Layman Explanation
Think of building a **small robot**:
- Step 1: Decide how many parts (layers) it will have.
- Step 2: Choose how it will learn (optimizer).
- Step 3: Give it practice problems (training).
- Step 4: Test how well it performs (evaluation).


```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
import numpy as np

# Step 1: Build the model
model = Sequential([
    Input(shape=(10,)),          # input layer with 10 features
    Dense(8, activation='relu'), # hidden layer
    Dense(1, activation='sigmoid') # output layer
])

# Step 2: Compile the model
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# Step 3: Dummy dataset
X = np.random.rand(200, 10)   # 200 samples, 10 features
y = np.random.randint(0, 2, (200, 1))  # binary labels

# Step 4: Train the model
history = model.fit(X, y, epochs=10, batch_size=16, verbose=1)

# Step 5: Evaluate the model
loss, accuracy = model.evaluate(X, y, verbose=0)
print("Model Accuracy:", accuracy)
```

- We built a robot (neural network) with 3 parts: input, hidden, output.
- We told it how to learn (Adam optimizer).
- We gave it practice problems (dummy dataset).
- It practiced for 10 rounds (epochs).
- Finally, we tested how well it learned (accuracy).

ItвЂ™s like teaching a student:
- Input = questions given.

- Output = studentвЂ™s answer.
