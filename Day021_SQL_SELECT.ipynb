```python
import sqlite3
import pandas as pd

# Connect to a new SQLite database file
conn = sqlite3.connect("Day21.db")
cursor = conn.cursor()

# Drop table if it already exists
cursor.execute("DROP TABLE IF EXISTS employees")

# Create employees table
cursor.execute("""
CREATE TABLE employees (
    emp_id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary INTEGER
)
""")
```

```python
# Employee data
data = [
    (1, 'Amit', 'IT', 60000),
    (2, 'Neha', 'HR', 55000),
    (3, 'Raj', 'Finance', 70000),
    (4, 'Simran', 'IT', 65000),
    (5, 'Karan', 'Marketing', 50000),
    (6, 'Priya', 'Finance', 72000),
    (7, 'Arjun', 'HR', 48000),
    (8, 'Meera', 'IT', 68000)
]

# Insert rows
cursor.executemany("INSERT INTO employees VALUES (?, ?, ?, ?)", data)
conn.commit()
```

```python
# Fetching ALL rows and ALL columns from the employees table
pd.read_sql("SELECT * FROM employees", conn)
```

```python
# Fetching only the name and salary columns
pd.read_sql("SELECT name, salary FROM employees", conn)
```

```python
# Renaming columns to make them more readable
pd.read_sql("SELECT name AS employee_name, salary AS monthly_salary FROM employees", conn)
```

```python
# Fetching unique department names (removes duplicates)
pd.read_sql("SELECT DISTINCT department FROM employees", conn)
```

```python
# Fetching only the first 3 rows from the table
pd.read_sql("SELECT * FROM employees LIMIT 3", conn)
```
