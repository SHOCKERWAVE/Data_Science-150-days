```python
import statistics as stats

# Day 17: Statistics Basics

numbers = [10, 20, 20, 30, 40, 50]

# Mean
mean_value = stats.mean(numbers)
print("Mean:", mean_value)

# Median
median_value = stats.median(numbers)
print("Median:", median_value)

# Mode
mode_value = stats.mode(numbers)
print("Mode:", mode_value)
```

```python
# Mean manually
numbers = [1, 2, 3, 4, 5]
mean_manual = sum(numbers) / len(numbers)
print("Manual Mean:", mean_manual)
```
