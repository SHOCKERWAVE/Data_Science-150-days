```python
# Day 16: Probability Basics

# Example 1: Coin toss
favorable = 1   # heads
total = 2       # heads or tails
prob_heads = favorable / total
print("Probability of heads:", prob_heads)

# Example 2: Dice roll
favorable = 1   # only one face = 6
total = 6       # six faces
prob_six = favorable / total
print("Probability of rolling a 6:", prob_six)

# Example 3: Probability of even number on dice
favorable = 3   # (2,4,6)
total = 6
prob_even = favorable / total
print("Probability of even number:", prob_even)
```
