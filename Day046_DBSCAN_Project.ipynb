```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
```

```python
# Example dataset: Credit card transactions (simplified)
# Goal: Detect normal vs suspicious spending patterns

data = {
    'TransactionAmount': [50,55,60,65,70,75,80,200,210,220,5,10,15,500,520,540],
    'TransactionFrequency': [5,6,7,8,9,10,11,2,3,4,20,21,22,1,1,2]
}
df = pd.DataFrame(data)

df.head()
```

```python
# eps = max distance for neighbors
# min_samples = minimum points to form a cluster
dbscan = DBSCAN(eps=15, min_samples=2)
df['Cluster'] = dbscan.fit_predict(df[['TransactionAmount','TransactionFrequency']])
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='TransactionAmount', y='TransactionFrequency', hue='Cluster', data=df, palette='Set1')
plt.xlabel("Transaction Amount ($)")
plt.ylabel("Transaction Frequency")
plt.title("DBSCAN Clustering of Transactions")
plt.show()
```
