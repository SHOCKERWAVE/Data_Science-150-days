```python
import pandas as pd

# Sample dataset with missing values
data = {
    "Name": ["Shubham", "Rahul", "Priya", "Amit"],
    "Age": [22, None, 23, None],
    "City": ["Prayagraj", "Delhi", None, "Mumbai"]
}

df = pd.DataFrame(data)
print("Original DataFrame:\n", df)
```

```python
# Check for missing values
print("\nMissing values:\n", df.isnull())

# Count missing values per column
print("\nCount of missing values:\n", df.isnull().sum())
```

```python
# Fill missing values with default
df_fill = df.fillna("Unknown")
print("\nAfter filling missing values:\n", df_fill)

# Fill numeric column with mean
df["Age"] = df["Age"].fillna(df["Age"].mean())
print("\nAfter filling Age with mean:\n", df)
```

```python
# Drop rows with missing values
df_drop = df.dropna()
print("\nAfter dropping rows with missing values:\n", df_drop)

# Drop columns with missing values
df_drop_col = df.dropna(axis=1)
print("\nAfter dropping columns with missing values:\n", df_drop_col)
```
