 Step 1 — What are Categorical Variables?
- Definition: Features that represent categories instead of numbers.
- Examples: Gender (Male/Female), City (Delhi, Mumbai, Bangalore).
- Layman analogy: Imagine a cricket team list with player names. Models don’t understand names directly — they need numbers.

 Step 2 — Encoding Techniques
- Label Encoding: Assigns each category a number.
- Example: Male = 0, Female = 1.
- Problem: Implies order (0 < 1), which may not be true.
- One‑Hot Encoding: Creates a new column for each category with 0/1 values.
- Example: City → Delhi [1,0,0], Mumbai [0,1,0], Bangalore [0,0,1].
-  Safer for non‑ordered categories.

```python
# Step 3
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Sample dataset
data = pd.DataFrame({
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Delhi', 'Mumbai'],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male']
})

print("Original Data:\n", data)

# Label Encoding for Gender
label_encoder = LabelEncoder()
data['Gender_encoded'] = label_encoder.fit_transform(data['Gender'])
print("\nAfter Label Encoding:\n", data)

# One-Hot Encoding for City
onehot = pd.get_dummies(data['City'], prefix='City')
data = pd.concat([data, onehot], axis=1)
print("\nAfter One-Hot Encoding:\n", data)
```

- Layman comments:
- Label Encoding = like giving roll numbers to students (Male=0, Female=1).
- One‑Hot Encoding = like making separate attendance sheets for each city.
