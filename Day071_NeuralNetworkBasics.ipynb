# Neural Network Basics

**Goal:** Learn the structure and working of a neural network.


A **neural network** is like a simplified version of the human brain.  
It has layers of "neurons" (tiny decision makers) that pass information forward.

---

## Structure
- **Input Layer** → takes in the data (like exam questions).
- **Hidden Layers** → process the data (students discussing answers).
- **Output Layer** → gives the final result (the final answer).

---

## Layman Explanation
Imagine a classroom:
- The teacher gives a question (input).
- Students discuss among themselves (hidden layers).
- The class leader announces the final answer (output).

Each student adjusts their confidence (weights) after feedback until the group learns to answer correctly.


```python
# Day 71 - Neural Network Basics

import numpy as np

# Example: simple feedforward neural network (conceptual)
def simple_nn(X, W1, W2):
    # X = input
    # W1, W2 = weights
    hidden = np.dot(X, W1)   # hidden layer computation
    output = np.dot(hidden, W2)  # output layer computation
    return output

# Dummy data
X = np.array([1, 2])  # input features
W1 = np.array([[0.5, 0.2], [0.3, 0.7]])  # weights for hidden layer
W2 = np.array([[0.6], [0.9]])            # weights for output layer

print("Output:", simple_nn(X, W1, W2))
```

### Key Ideas:
- **Weights:** Numbers that show how important each input is.
- **Bias:** Extra help to shift the answer.
- **Activation Function:** Decides if a neuron should "fire" or stay quiet.
- **Learning:** Adjusting weights after mistakes.

### Layman:
- Weights = how much each student trusts an idea.
- Bias = a hint to help.
- Activation = deciding if the idea is strong enough to share.
- Learning = correcting mistakes after feedback.
