# Transformers Basics

**Goal:** Learn the core idea behind transformer models and why they are powerful for NLP and generative AI.


Transformers are deep learning models designed for sequence data (like text).  
They use **attention mechanisms** to focus on important parts of input.

---

## Key Features
- **Self-Attention** → Looks at all words in a sentence and decides which are most important.
- **Parallel Processing** → Processes words simultaneously (faster than RNNs).
- **Scalability** → Works well for very large datasets.

---

## Layman Explanation
Imagine reading a book:
- Instead of reading word by word (like RNNs), transformers look at the **whole page at once**.
- They highlight important words (attention).
- This makes them faster and smarter at understanding context.


```python
import tensorflow as tf
from tensorflow.keras.layers import MultiHeadAttention, LayerNormalization, Dense, Input
from tensorflow.keras.models import Model

# Input
inputs = Input(shape=(None, 64))  # sequence length, embedding size

# Self-Attention
attention = MultiHeadAttention(num_heads=2, key_dim=64)(inputs, inputs)

# Add & Normalize
x = LayerNormalization(epsilon=1e-6)(inputs + attention)

# Feed Forward
ff = Dense(64, activation='relu')(x)
outputs = Dense(64)(ff)

# Build model
transformer_block = Model(inputs=inputs, outputs=outputs)
transformer_block.summary()
```

- Input = sequence of word embeddings.
- MultiHeadAttention = looks at relationships between words.
- LayerNormalization = keeps values stable.
- Feed Forward = processes information further.

Layman:
It’s like a student reading a paragraph:
- Attention = highlighting key words.
- Normalization = keeping notes neat.
- Feed Forward = thinking deeper about the meaning.
