# Embeddings

**Goal:** Understand embeddings and how they represent words or data as numerical vectors.


**Embeddings** are numerical representations of data (like words or images).  
They capture meaning and relationships in a way that machines can process.

---

## Why Embeddings?
- Convert words into numbers for neural networks.
- Preserve meaning (similar words have similar vectors).
- Enable tasks like search, translation, and text generation.

---

## Layman Explanation
Think of embeddings as a **dictionary for the robot**:
- Each word is translated into a unique number pattern.
- Similar words (like "king" and "queen") have similar patterns.
- This helps the robot understand context instead of just memorizing words.


```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding
import numpy as np

# Example: 1000 words in vocabulary, each mapped to 64-dimensional vector
embedding_layer = Embedding(input_dim=1000, output_dim=64)

# Dummy input: sequence of word indices
word_indices = np.array([[1, 5, 42, 999]])

# Get embeddings
embeddings = embedding_layer(word_indices)
print("Embeddings shape:", embeddings.shape)
```

- We created an embedding layer that maps words to 64-dimensional vectors.
- Input = word indices (like dictionary numbers).
- Output = embeddings (unique number patterns for each word).

Layman:
It’s like giving each word a **barcode**:
- Input = word ID.
- Output = barcode that captures meaning.
- Similar words have similar barcodes.
