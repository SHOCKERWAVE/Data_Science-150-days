```python
# Day 4 - Python If / Else

temp = 35

# if = if this is true, then do this
if temp > 30:
    print("It's warm")

# elif = if the first condition is false, try this next condition
elif temp > 20:
    print("The weather is pleasant.")

# else = if both conditions above are false, do this
else:
    print("It's cold outside.")
```

    # Day 4 - Python If / Else
    Learned:
    - If = check a condition
    - Else = run when condition is false
    - Elif = check more conditions
