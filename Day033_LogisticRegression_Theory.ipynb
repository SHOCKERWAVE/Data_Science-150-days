 Step 1 — What is Logistic Regression?
- Definition (simple words): Logistic regression is used when the output is categorical (yes/no, true/false, 0/1).
- Instead of predicting a continuous number (like salary), it predicts the probability of belonging to a class.
 Layman analogy: Imagine predicting whether a student passes or fails based on study hours. The answer is not a number like “65 marks,” but a category: Pass (1) or Fail (0).

 Step 2 — Logistic Function (Sigmoid)
The core of logistic regression is the sigmoid function:
P(Y=1|X)=\frac{1}{1+e^{-(\beta _0+\beta _1X)}}
- Output is always between 0 and 1 (probability).
- If probability > 0.5 → Predict class 1 (Yes).
- If probability ≤ 0.5 → Predict class 0 (No).

 Step 3 — Key Concepts
- Odds: Ratio of probability of success to failure.
- Logit: Natural log of odds. Logistic regression models the logit as a linear function of X.
- Decision boundary: The threshold (usually 0.5) that decides class.

 Step 4 — Assumptions
- Dependent variable is categorical.
- Observations are independent.
- No extreme multicollinearity among predictors.
- Large sample size preferred.

 Step 5 — Advantages
- Works well for binary classification.
- Outputs probabilities, not just labels.
- Easy to interpret coefficients.

 Step 6 — Limitations
- Only works for linear decision boundaries.
- Not suitable for complex non‑linear problems.
- Sensitive to outliers.
