```python
# Day 1: Variables & Data Types

# Think of variables as boxes where you store information.
name = "Shubham"                  # a text box
age = 22                          # a number box
height = 5.11                     # a decimal number box
is_student = True                 # a true/false box

# Show what's inside the boxes
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Student:", is_student)

# Change the label on the box (conver number to text)
age_str = str(age)
print("Age as string:", age_str)
```
